```python
# (1) Case of alpha == 0.5

def get_initials(s):
    s = s.strip().split()
    s = map(lambda x: x[0], s)
    return ''.join(s)


case_1 = df \
    .query('num_workers >= 4') \
    [['agg_method', 'num_workers', 'agg_prob', 'elastic_alpha', 'rank_0.TestAcc', 'exp_id', 'agg.AggTestAcc']] \
    .query('(agg_method != "gossipingSgd") or (agg_prob < 0.5)') \
    .query('(elastic_alpha == 0.5) or (agg_method != "elasticGossip")') \
    .sort_values(['num_workers', 'agg_prob', 'elastic_alpha'], ascending=[True, False, True]) \
    .assign(elastic_alpha=lambda x: 
            x.apply(lambda r: r.elastic_alpha if r.agg_method == 'elasticGossip' else None, axis=1)) \
    .assign(agg_prob=lambda x:
            x.apply(lambda r: r.agg_prob if (r.agg_method in ['elasticGossip', 'gossipingSgd']) else None, axis=1)) \
    .assign(agg_method=lambda x:
            x.agg_method.map({
                'gradAllReduce': 'All Reduce',
                'noComm': 'No Communication',
                'elasticGossip': 'Elastic Gossip',
                'gossipingSgd': 'Gossiping SGD',
            })) \
    .rename(columns={'rank_0.TestAcc': 'Rank-0 Accuracy',
                     'agg.AggTestAcc': 'Aggregate Accuracy',
                     'agg_method': 'Method', 
                     'agg_prob': 'p',
                     'num_workers': 'W',
                     'elastic_alpha': 'alpha'}) \
    .assign(id=lambda d:
            d.apply(lambda x: '-'.join([
                get_initials(x['Method']),
                str(x['W']) +
                ('-%.3f' % x['p'] if x['p'] == x['p'] else ''),
            ]), axis=1)) \
    .reset_index(drop=True)

print (case_1 \
           [['Method', 'W', 'p', 'id', 'Rank-0 Accuracy', 'Aggregate Accuracy']] \
           .to_latex(index=False, na_rep='-'))

display(case_1[['exp_id', 'id']].values.tolist())
```

```python
# (2) Comparing communication probability with communication period

case_2 = df \
    .query('num_workers == 4') \
    .query('agg_method == "gossipingSgd"') \
    .query('agg_prob != 0.5') \
    [['agg_prob', 'agg_period', 'rank_0.TestAcc', 'exp_id', 'agg.AggTestAcc']]
case_2['agg_period'] = case_2['agg_period'].apply(lambda x: x if x > 1 else np.nan)
case_2['agg_prob'] = case_2['agg_prob'].apply(lambda x: x if x < 1 else np.nan)
case_2['effective_agg_period'] = case_2['agg_prob'].apply(lambda p: int(1/p) if p == p else p)
case_2 = case_2.rename(columns={
    'rank_0.TestAcc': 'Rank-0 Accuracy',
    'agg.AggTestAcc': 'Aggregate Accuracy',
    'agg_period': 'tau', 
    'effective_agg_period': 'tau_eff',
    'agg_prob': 'p',
})
case_2 = case_2.sort_values(by=['tau', 'tau_eff'])
display(case_2[['p', 'tau_eff', 'tau', 'Rank-0 Accuracy', 'Aggregate Accuracy']])
print (case_2 \
           [['p', 'tau_eff', 'tau', 'Rank-0 Accuracy', 'Aggregate Accuracy']] \
           .to_latex(index=False, na_rep='-'))
```

```python
# (3) Impact of alpha

case_3 = df \
    .query('agg_prob != 0.5') \
    .query('agg_prob == 0.00048828125 or agg_prob == 0.031250') \
    .query('num_workers != 8 or agg_prob != 0.031250') \
    .query('num_workers >= 4') \
    .query('agg_method == "elasticGossip"') \
    [['exp_id', 'num_workers', 'agg_prob', 'elastic_alpha', 'rank_0.TestAcc', 'agg.AggTestAcc']] \
    .sort_values(['num_workers', 'elastic_alpha']) \
    .rename(columns={
        'rank_0.TestAcc': 'Rank-0 Accuracy',
        'agg.AggTestAcc': 'Aggregate Accuracy',
        'agg_prob': 'p',
        'num_workers': 'W',
        'elastic_alpha': 'alpha',
    }) \
    .sort_values(['W', 'p', 'alpha'], ascending=[True, False, True]) \
    .assign(Label=lambda fr: fr.apply(lambda x: 'EG-' + str(x['W']) + '-%.4f' % x['p'] + '-%.2f' % x['alpha'], axis=1)) \
    [['exp_id', 'W', 'p', 'alpha', 'Label', 'Rank-0 Accuracy', 'Aggregate Accuracy']] \
    .reset_index(drop=True)

display(case_3)
print(case_3.drop(columns=['exp_id']).to_latex(index=False, na_rep='-'))
display(case_3[['exp_id', 'Label']].values.tolist())
```
